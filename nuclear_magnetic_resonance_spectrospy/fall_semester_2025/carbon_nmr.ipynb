# üß¨ Carbon-13 NMR Spectroscopy Analysis

## üìä Introduction to ¬π¬≥C-NMR

**Carbon-13 Nuclear Magnetic Resonance (¬π¬≥C-NMR)** is a powerful analytical technique that reveals the **carbon skeleton** of organic molecules. Unlike proton NMR, which shows hydrogen environments, ¬π¬≥C-NMR directly observes carbon nuclei, providing complementary structural information.

---

## ‚öõÔ∏è Nuclear Properties of Carbon-13

| Property | ¬π¬≥C | ¬πH (comparison) |
|----------|-----|-----------------|
| **Natural Abundance** | 1.11% | 99.985% |
| **Spin** | I = ¬Ω | I = ¬Ω |
| **Gyromagnetic Ratio (Œ≥)** | 10.71 MHz/T | 42.58 MHz/T |
| **Sensitivity (relative to ¬πH)** | 0.016 | 1.000 |
| **Typical Spectral Width** | 0-220 ppm | 0-12 ppm |

**Key Insight:** ¬π¬≥C-NMR is **~6,000√ó less sensitive** than ¬πH-NMR due to:
1. Lower natural abundance (1.11% vs 99.99%)
2. Smaller gyromagnetic ratio ($\gamma_{\text{C}}/\gamma_{\text{H}} \approx 0.25$)
3. Combined effect: $(0.0111)^2 \times (0.25)^4 \approx 1.6 \times 10^{-5}$

---

## üìê Chemical Shift in ¬π¬≥C-NMR

The **chemical shift** quantifies how far a nucleus resonates from a reference standard (TMS = tetramethylsilane):

$$
\delta = \frac{\nu_{\text{sample}} - \nu_{\text{TMS}}}{\nu_{\text{spectrometer}}} \times 10^6 \quad \text{(in ppm)}
$$

where:
- $\nu_{\text{sample}}$: Resonance frequency of carbon nucleus
- $\nu_{\text{TMS}}$: Resonance frequency of TMS reference (0 ppm)
- $\nu_{\text{spectrometer}}$: Operating frequency for ¬π¬≥C (typically 100 MHz)

---

## üé® Chemical Shift Ranges

Carbon chemical shifts span a **much wider range** (0-220 ppm) than proton shifts (0-12 ppm), providing excellent separation between different carbon types:

### **Aliphatic Carbons (sp¬≥ hybridization)**

$$
\text{sp}^3 \text{ C-H: } 0\text{-}50 \text{ ppm}
$$

| Carbon Type | Œ¥ (ppm) | Examples |
|-------------|---------|----------|
| CH‚ÇÉ (methyl) | 8-30 | -CH‚ÇÉ |
| CH‚ÇÇ (methylene) | 20-40 | -CH‚ÇÇ- |
| CH (methine) | 25-50 | >CH- |
| Quaternary C (R‚ÇÑC) | 30-50 | C(CH‚ÇÉ)‚ÇÑ |

**Shielding:** Alkyl carbons are **highly shielded** by surrounding electron density.

---

### **Heteroatom-Bonded Carbons**

$$
\text{C-O, C-N: } 50\text{-}90 \text{ ppm}
$$

**Electronegativity Effect:** Oxygen and nitrogen **withdraw electron density**, deshielding the carbon and shifting resonance downfield.

Examples:
- **Alcohols, Ethers:** C-O at 50-90 ppm
- **Amines:** C-N at 40-60 ppm
- **Carbohydrates:** Multiple C-O bonds ‚Üí 60-110 ppm

---

### **Unsaturated Carbons (sp¬≤ and sp hybridization)**

$$
\text{sp}^2 \text{ C=C: } 100\text{-}150 \text{ ppm} \quad | \quad \text{sp C‚â°C: } 65\text{-}85 \text{ ppm}
$$

| Carbon Type | Œ¥ (ppm) | Hybridization |
|-------------|---------|---------------|
| **Alkene** (C=C) | 100-150 | sp¬≤ |
| **Aromatic** (Ar-C) | 110-160 | sp¬≤ |
| **Alkyne** (C‚â°C) | 65-85 | sp |
| **Nitrile** (C‚â°N) | 115-125 | sp |

**œÄ-Electron Effects:** Unsaturated carbons experience **deshielding** from œÄ-electron circulation.

---

### **Carbonyl Carbons (C=O)**

$$
\text{C=O: } 160\text{-}220 \text{ ppm}
$$

**Most Deshielded:** Carbonyl carbons appear farthest downfield due to:
- sp¬≤ hybridization
- Highly electronegative oxygen
- Resonance structures

| Functional Group | Œ¥ (ppm) | Structure |
|-----------------|---------|-----------|
| **Ester, Amide** | 160-180 | R-CO-OR', R-CO-NR‚ÇÇ |
| **Ketone, Aldehyde** | 180-215 | R-CO-R', R-CHO |
| **Carboxylic Acid** | 170-185 | R-COOH |

---

## üî¨ Key Differences from ¬πH-NMR

### **1. Coupling Patterns**

**¬π¬≥C-¬π¬≥C Coupling is Negligible:**

$$
P(\text{two adjacent } ^{13}\text{C}) = (0.0111)^2 \approx 0.01\%
$$

Unlike ¬πH-NMR where vicinal coupling creates multiplets, ¬π¬≥C-NMR peaks are typically **singlets** (in decoupled mode).

### **2. Proton Decoupling**

Most ¬π¬≥C spectra are acquired with **broadband proton decoupling**:

$$
^{13}\text{C}\{^{1}\text{H}\} \quad \text{(notation for decoupled spectrum)}
$$

**Effect:** Eliminates ¬π¬≥C-¬πH coupling, collapsing multiplets into **single sharp peaks**.

**Advantages:**
- ‚úÖ Simplified spectra (one peak per unique carbon)
- ‚úÖ Increased sensitivity (signal not split into multiplets)
- ‚úÖ Easier peak assignment

### **3. Integration Limitations**

**‚ö†Ô∏è Quantitative analysis is difficult** in ¬π¬≥C-NMR because:
- Different carbons have different **relaxation times** (T‚ÇÅ)
- **Nuclear Overhauser Effect (NOE)** varies with carbon type
- Signal intensity ‚â† number of carbons

**Solution:** Use specialized techniques (inverse-gated decoupling) for quantitative work.

---

## üßÆ Spectrometer Frequency Calculation

For a magnet where ¬πH resonates at 400 MHz, ¬π¬≥C frequency is calculated from gyromagnetic ratios:

$$
\nu_{^{13}\text{C}} = \frac{\gamma_{^{13}\text{C}}}{\gamma_{^{1}\text{H}}} \times \nu_{^{1}\text{H}} = \frac{10.71}{42.58} \times 399.78 \text{ MHz} = 100.52 \text{ MHz}
$$

**Common ¬π¬≥C Frequencies:**
- 100 MHz (at 400 MHz ¬πH field)
- 125 MHz (at 500 MHz ¬πH field)
- 150 MHz (at 600 MHz ¬πH field)

---

## üéØ Applications of ¬π¬≥C-NMR

1. **Structure Elucidation:** Determine carbon framework and functional groups
2. **Stereochemistry:** Distinguish diastereomers by subtle chemical shift differences
3. **Dynamics:** Study molecular motion and conformational exchange
4. **Natural Products:** Identify complex organic molecules from biological sources
5. **Polymer Analysis:** Characterize repeat units, tacticity, and chain ends

---

üí° **Summary:** ¬π¬≥C-NMR provides the **carbon skeleton blueprint** of molecules, complementing ¬πH-NMR's hydrogen connectivity information. Together, they form the cornerstone of modern structure determination.

## üß† Deep Learning Signal Enhancement

### üéØ Why Denoise Carbon-13 FIDs?

Carbon-13 NMR suffers from **inherently low sensitivity** due to:

1. **Low Natural Abundance:** Only 1.11% of carbon atoms are ¬π¬≥C
2. **Small Gyromagnetic Ratio:** Œ≥_C ‚âà ¬º of Œ≥_H
3. **Combined Effect:** Signal is **~6,000√ó weaker** than ¬πH-NMR

**Traditional Solutions:**
- ‚ùå More scans ‚Üí Longer acquisition time (SNR ‚àù ‚àöN)
- ‚ùå Stronger magnet ‚Üí Expensive ($500K+ upgrade)
- ‚ùå More sample ‚Üí Often limited/precious

**Deep Learning Solution:**
- ‚úÖ Train neural network to remove noise while preserving signal
- ‚úÖ Equivalent to 4-16√ó more scans (SNR improvement of 6-12 dB)
- ‚úÖ Processing time: ~5 ms per FID

---

### üèóÔ∏è Denoising Architecture

We use **DenoiseNetPhysics**, a dilated residual CNN optimized for NMR:

$$
\text{FID}_{\text{clean}} = \text{FID}_{\text{noisy}} + f_\theta(\text{FID}_{\text{noisy}})
$$

**Key Features:**
- **Residual connections:** Preserves original signal, only adds corrections
- **Dilated convolutions:** Captures long-range T‚ÇÇ decay patterns
- **Complex-valued processing:** Handles (Real, Imag) channels simultaneously

---

### üìä Expected SNR Improvement

$$
\Delta\text{SNR} = 20 \log_{10}\left(\frac{\text{RMS}_{\text{signal,denoised}}}{\text{RMS}_{\text{noise,denoised}}} \div \frac{\text{RMS}_{\text{signal,original}}}{\text{RMS}_{\text{noise,original}}}\right)
$$

**Typical Results:**
- Good quality FID: +4 to +6 dB
- Low SNR FID: +6 to +10 dB
- Already clean FID: +1 to +3 dB

**Acquisition Time Equivalent:**

$$
\Delta\text{SNR} = 6 \text{ dB} \equiv \sqrt{N_{\text{scans}}} \times 2 \quad \Rightarrow \quad N_{\text{scans}} = 4\times \text{ (16√ó time savings)}
$$

---

### ‚öôÔ∏è How It Works

**Step 1:** Normalize FID using noise floor detection  
**Step 2:** Feed (Real, Imaginary) components into neural network  
**Step 3:** Model outputs denoised signal preserving peak positions  
**Step 4:** Calculate SNR improvement for validation

**Important:** This cell is **optional** but recommended for low-SNR ¬π¬≥C data. Set `USE_DENOISING = False` to skip.

````python
# SET THIS FLAG TO ENABLE/DISABLE DENOISING
USE_DENOISING = True  # Set to False to skip denoising

if USE_DENOISING:
    import torch
    import torch.nn as nn
    import torch.nn.functional as F
    
    # Define the denoiser architecture
    class DilatedResBlock(nn.Module):
        def __init__(self, channels: int, dilation: int = 1, k: int = 11):
            super().__init__()
            pad = dilation * ((k - 1) // 2)
            self.conv1 = nn.Conv1d(channels, channels, k, padding=pad, dilation=dilation)
            self.conv2 = nn.Conv1d(channels, channels, k, padding=pad, dilation=dilation)
        def forward(self, x):
            h = F.relu(self.conv1(x))
            h = self.conv2(h)
            return F.relu(x + h)
    
    class DenoiseNetPhysics(nn.Module):
        def __init__(self, in_ch=2, hidden=64, k=11, dilations=(1,2,4,8,16,32)):
            super().__init__()
            self.inp = nn.Conv1d(in_ch, hidden, k, padding=(k-1)//2)
            self.blocks = nn.Sequential(*[DilatedResBlock(hidden, d, k) for d in dilations])
            self.out = nn.Conv1d(hidden, in_ch, k, padding=(k-1)//2)
        def forward(self, x):
            h = F.relu(self.inp(x))
            h = self.blocks(h)
            return x + self.out(h)
    
    # Load model
    device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
    denoiser_model = DenoiseNetPhysics().to(device)
    
    # Try loading checkpoint
    checkpoint_dir = Path("checkpoints")
    best_models = ["DenoiseNetPhysics_final_best.pth", "DenoiseNetPhysics_aggressive_best.pth",
                   "DenoiseNetPhysics_extended_best.pth", "DenoiseNetPhysics_finetuned_best.pth"]
    
    loaded = False
    for ckpt_name in best_models:
        ckpt_path = checkpoint_dir / ckpt_name
        if ckpt_path.exists():
            try:
                ckpt = torch.load(ckpt_path, map_location=device, weights_only=False)
                denoiser_model.load_state_dict(ckpt['model_state'] if 'model_state' in ckpt else ckpt)
                print(f"‚úÖ Loaded: {ckpt_name}")
                loaded = True
                break
            except: continue
    
    if loaded:
        denoiser_model.eval()
        # Apply denoising
        data_original = data.copy()
        real_part, imag_part = data[:, 1].astype(np.float32), data[:, 2].astype(np.float32)
        sig = np.stack([real_part, imag_part])
        
        # Normalize
        L = sig.shape[1]
        late = slice(int(0.8*L), L)
        baseline = np.median(sig[:, late], axis=1, keepdims=True)
        sig_norm = sig - baseline
        
        # Denoise
        with torch.no_grad():
            x = torch.from_numpy(sig_norm).unsqueeze(0).to(device)
            y = denoiser_model(x).cpu().squeeze(0).numpy()
        
        # Update data
        data[:, 1] = (y[0] + baseline[0, 0])
        data[:, 2] = (y[1] + baseline[1, 0])
        
        # Calculate SNR
        def calc_snr(d):
            env = np.sqrt(d[:,1]**2 + d[:,2]**2)
            s = int(0.2*len(env))
            return 20*np.log10(np.sqrt(np.mean(env[:s]**2)) / (np.sqrt(np.mean(env[-s:]**2))+1e-12))
        
        snr_before, snr_after = calc_snr(data_original), calc_snr(data)
        snr_improvement = snr_after - snr_before
        print(f"   SNR: {snr_before:.1f} ‚Üí {snr_after:.1f} dB ({snr_improvement:+.1f} dB)")
        print("‚úÖ Using DENOISED data for analysis")
    else:
        print("‚ö†Ô∏è  No model loaded - using original data")
        USE_DENOISING = False
else:
    print("Denoising disabled - using original FID")
````

## üåä Gaussian Smoothing: Frequency-Domain Filtering

### üìê Mathematical Foundation

After FFT converts the FID to a spectrum, we can optionally apply **Gaussian smoothing** to reduce residual high-frequency noise. This is a **post-processing** technique complementary to deep learning denoising.

---

### üßÆ The Gaussian Filter

A **Gaussian filter** convolves the spectrum with a Gaussian kernel:

$$
G(x) = \frac{1}{\sigma\sqrt{2\pi}} \exp\left(-\frac{x^2}{2\sigma^2}\right)
$$

**Discrete 1D Convolution:**

$$
S_{\text{smoothed}}[i] = \sum_{j=-\infty}^{\infty} S_{\text{original}}[j] \cdot G(i - j)
$$

where $\sigma$ (sigma) controls the **smoothing width**:
- **Small œÉ (1-2):** Gentle smoothing, preserves peak shape
- **Medium œÉ (3-5):** Moderate smoothing, slight peak broadening
- **Large œÉ (>5):** Aggressive smoothing, significant peak broadening

---

### üìä Effect on NMR Spectra

**Frequency Domain Application:**

$$
\text{Spectrum}_{\text{smooth}} = \text{Gaussian}(\sigma) * \text{Spectrum}_{\text{original}}
$$

**Advantages:**
- ‚úÖ Reduces **high-frequency noise** (baseline fluctuations)
- ‚úÖ Improves **visual clarity** of peaks
- ‚úÖ Smooths out **FFT artifacts** (truncation effects)

**Disadvantages:**
- ‚ö†Ô∏è **Broadens peaks** (reduces resolution)
- ‚ö†Ô∏è **Reduces peak height** (conservation of area)
- ‚ö†Ô∏è Can **merge close peaks** if œÉ too large

---

### üî¨ When to Use Gaussian Smoothing?

**Decision Tree:**

```
If (SNR improvement from denoising) >= +2 dB:
    ‚Üí Skip Gaussian smoothing (signal already clean)
Else:
    ‚Üí Apply Gaussian smoothing (œÉ = 1-3)
```

**Rationale:**
- Deep learning denoising works in **time domain** (preserves peak shape)
- Gaussian smoothing works in **frequency domain** (may distort peaks)
- **Best practice:** Use DL denoising as primary method, Gaussian as backup

---

### ‚öôÔ∏è Implementation

Using `scipy.ndimage.gaussian_filter1d`:

$$
\text{filtered} = \text{gaussian\_filter1d}(\text{spectrum}, \sigma=\sigma_{\text{value}})
$$

**Default Parameters:**
- `sigma = 2.0` for ¬π¬≥C-NMR (balance between noise reduction and resolution)
- `mode = 'reflect'` (boundary condition to avoid edge artifacts)

---

### üìà Recommended œÉ Values

| Spectrum Quality | Recommended œÉ | Effect |
|-----------------|---------------|--------|
| **High SNR** (>30 dB) | 0 (skip) | Preserve original resolution |
| **Medium SNR** (20-30 dB) | 1-2 | Minimal smoothing |
| **Low SNR** (<20 dB) | 2-4 | Moderate smoothing |
| **Very Noisy** (<10 dB) | 3-5 | Aggressive smoothing |

---

### üéØ Comparison with Apodization

| Method | Domain | Effect | Peak Shape |
|--------|--------|--------|------------|
| **Apodization** | Time (FID) | Exponential weighting | Lorentzian broadening |
| **Gaussian Filter** | Frequency (Spectrum) | Convolution smoothing | Gaussian broadening |

**Apodization (Time Domain):**

$$
\text{FID}_{\text{apod}}(t) = \text{FID}(t) \cdot e^{-t/T_{\text{apod}}}
$$

**Gaussian (Frequency Domain):**

$$
\text{Spectrum}_{\text{smooth}}(\nu) = \text{Spectrum}(\nu) * G_\sigma(\nu)
$$

## üî¨ Fourier Transform: Time ‚Üí Frequency Conversion

### üìê The Discrete Fourier Transform (DFT)

NMR spectroscopy relies on converting the **time-domain FID** into a **frequency-domain spectrum** using the Fourier Transform:

$$
S(\omega) = \int_{-\infty}^{\infty} \text{FID}(t) \cdot e^{-i\omega t} \, dt
$$

For **discretely sampled** NMR data (N points), we use the **Discrete Fourier Transform (DFT)**:

$$
S[k] = \sum_{n=0}^{N-1} \text{FID}[n] \cdot e^{-i2\pi kn/N}
$$

where:
- $\text{FID}[n]$ = Time-domain signal at point $n$
- $S[k]$ = Frequency-domain spectrum at point $k$
- $N$ = Total number of data points

---

### ‚ö° Fast Fourier Transform (FFT)

The **FFT** is an efficient algorithm to compute the DFT:

**Complexity:**
- DFT (naive): $O(N^{2})$ operations
- **FFT (Cooley-Tukey):** $O(N \log N)$ operations

**Speed Comparison:**
- For $N = 16,384$ points (typical FID):
  - DFT: ~268 million operations
  - **FFT: ~230,000 operations** (1,160√ó faster!)

**NMR Processing:** All modern NMR software uses FFT for real-time spectral display.

---

### üéµ Spectral Width and Nyquist Frequency

The **spectral width (SW)** is determined by the **dwell time** ($\Delta t$) between FID points:

$$
\text{SW} = \frac{1}{\Delta t} \quad \text{(in Hz)}
$$

**Nyquist Theorem:** Maximum observable frequency is:

$$
f_{\text{Nyquist}} = \frac{\text{SW}}{2} = \frac{1}{2\Delta t}
$$

**Example:**
- $\Delta t = 50$ Œºs ‚Üí SW = 20,000 Hz ‚Üí $f_{\text{Nyquist}} = 10,000$ Hz (10 kHz)

---

### üßÆ ¬π¬≥C Spectrometer Frequency

For a magnet where ¬πH resonates at 399.78 MHz, the ¬π¬≥C operating frequency is:

$$
\nu_{^{13}\text{C}} = \frac{\gamma_{^{13}\text{C}}}{\gamma_{^{1}\text{H}}} \times \nu_{^{1}\text{H}}
$$

$$
\nu_{^{13}\text{C}} = \frac{10.71 \text{ MHz/T}}{42.58 \text{ MHz/T}} \times 399.78 \text{ MHz} = 100.52 \text{ MHz}
$$

**Gyromagnetic Ratio:**

$$
\gamma = \frac{2\pi \cdot \nu_{0}}{B_{0}} \quad \text{(rad} \cdot \text{s}^{-1} \cdot \text{T}^{-1}\text{)}
$$

---

### üìä Chemical Shift Conversion

Convert **frequency** (Hz) to **chemical shift** (ppm):

$$
\delta_{\text{ppm}} = \frac{\nu_{\text{Hz}}}{\nu_{\text{spectrometer}}} = \frac{\nu_{\text{Hz}}}{100.52 \text{ MHz}}
$$

**Example:**
- Peak at 10,000 Hz ‚Üí $\delta = 10,000 / 100.52 = 99.5$ ppm (aromatic carbon)

---

### üåä Windowing (Apodization)

Before FFT, apply an **exponential window** to reduce truncation artifacts:

$$
\text{FID}_{\\text{windowed}}(t) = \text{FID}(t) \\cdot e^{-t/T_{\\text{window}}}
$$

**Effect:** Trades **resolution** for **sensitivity** (broader peaks, better SNR)

---

# Peak Detection with Carbon-13 Specific Thresholds

Carbon-13 NMR peaks are typically:
- **Weaker** than proton NMR (due to lower natural abundance)
- **Sharper** when proton-decoupled
- **More widely spaced** (larger ppm range)

We'll use adjusted thresholds for peak detection suitable for Carbon-13 spectra.

````python
def plot_full_carbon_and_zoom_with_peaks(
    frequencies,
    magnitudes,
    title,
    spectrometer_freq=399.78219838,
    ppm_range=None,
    identify_functional_groups=None,
    ppm_shifts=None,
    identified_groups=None,
    min_distance_hz=7.0,
    height_frac=0.10,
    prominence_frac=0.05,
    ax_full=None,
    ax_zoom=None,
    show=True
):
    # Rename parameters to internal variables
    freqs = frequencies
    mags = magnitudes
    
    # Determine if full spectrum (has negatives) or only positive
    is_full = (freqs.min() < 0) and (freqs.max() > 0)

    if is_full:
        freqs_shifted = fftshift(freqs)
        mags_shifted = fftshift(mags)
    else:
        freqs_shifted = freqs
        mags_shifted = mags

    # Compute ppm axis
    ppm_axis = freqs_shifted / spectrometer_freq

    # Infer ppm_range (do NOT clamp for 13C; caller can pass explicit range)
    if ppm_range is None:
        ppm_range = float(ppm_axis.max() - ppm_axis.min())

    # Peak detection thresholds
    max_int = float(mags_shifted.max())
    if max_int == 0:
        print("All-zero intensity.")
        return {}

    height = height_frac * max_int
    prominence = prominence_frac * max_int

    n_points = len(mags_shifted)
    # Use actual frequency span (Hz) to compute Hz-per-point robustly
    hz_span = float(freqs_shifted.max() - freqs_shifted.min()) if n_points > 1 else spectrometer_freq
    hz_per_point = hz_span / max(1, (n_points - 1))
    min_distance_pts = max(1, int(min_distance_hz / max(hz_per_point, 1e-9)))

    from scipy.signal import find_peaks
    peaks, properties = find_peaks(
        mags_shifted,
        height=height,
        distance=min_distance_pts,
        prominence=prominence
    )

    # Functional group identification (expects ppm positions, not frequencies)
    if identified_groups is None and callable(identify_functional_groups) and ppm_shifts is not None:
        # Convert to ppm positions for identification
        # Re-run identification on ppm positions at detected peaks
        try:
            identified_groups = identify_functional_groups(ppm_axis, mags_shifted, ppm_shifts)
        except Exception:
            identified_groups = []
    identified_groups = identified_groups or []

    # Build zoom window:
    # Prefer detected peaks; if absent, fall back to identified_groups.
    buffer_ppm = 0.10
    zoom_x_min, zoom_x_max = None, None
    if peaks.size:
        peak_ppms = ppm_axis[peaks]
        zoom_x_min = float(peak_ppms.min() - buffer_ppm)
        zoom_x_max = float(peak_ppms.max() + buffer_ppm)
        print(f"Zoom window from detected peaks: {zoom_x_min:.2f}‚Äì{zoom_x_max:.2f} ppm")
    elif identified_groups:
        group_ppms = [p for p, _ in identified_groups]
        zoom_x_min = float(min(group_ppms) - buffer_ppm)
        zoom_x_max = float(max(group_ppms) + buffer_ppm)
        print(f"Zoom window from functional groups: {zoom_x_min:.2f}‚Äì{zoom_x_max:.2f} ppm")
    else:
        # Default to whole ppm axis
        zoom_x_min = float(ppm_axis.min())
        zoom_x_max = float(ppm_axis.max())
        print("No peaks/groups detected; zoom window = full ppm axis.")

    created_fig = False
    if ax_full is None or ax_zoom is None:
        fig, (ax_full, ax_zoom) = plt.subplots(1, 2, figsize=(14, 5), sharey=True)
        created_fig = True
    else:
        fig = ax_full.figure

    # Full plot (entire spectrum, like H-NMR cell 25)
    ax_full.plot(ppm_axis, mags_shifted, label="Spectrum (FFT)")
    if peaks.size:
        ax_full.plot(ppm_axis[peaks], mags_shifted[peaks], "x", label="Detected Peaks")
    ax_full.invert_xaxis()
    ax_full.set_xlabel("Chemical Shift (ppm)")
    ax_full.set_ylabel("Intensity (A.U.)")
    ax_full.set_title(title)
    ax_full.legend()

    # Zoom plot
    ax_zoom.plot(ppm_axis, mags_shifted, label="Spectrum (FFT)")
    if peaks.size:
        ax_zoom.plot(ppm_axis[peaks], mags_shifted[peaks], "x", label="Detected Peaks")
    ax_zoom.invert_xaxis()
    ax_zoom.set_xlim(zoom_x_max, zoom_x_min)  # show all detected peaks
    ax_zoom.set_xlabel("Chemical Shift (ppm)")
    ax_zoom.set_title(f"{title} (zoom)")
    ax_zoom.legend()

    if show and created_fig:
        plt.tight_layout()
        plt.show()
    elif (not show) and created_fig:
        # Suppress auto-display in notebooks by closing the figure
        plt.close(fig)

    return {
        "fig": fig,
        "axes": (ax_full, ax_zoom),
        "ppm_axis": ppm_axis,
        "intensity": mags_shifted,
        "peaks": peaks,
        "properties": properties,
        "identified_groups": identified_groups,
    }


# Carbon-13 specific peak detection parameters
# Adjusted to detect peaks including weak ones near 0 ppm (e.g., TMS region)
C13_HEIGHT_FRAC = 0.03      # Lower threshold to catch weaker peaks (was 0.05)
C13_PROMINENCE_FRAC = 0.025 # Lower prominence to detect smaller peaks (was 0.04)
C13_MIN_DISTANCE_HZ = 10.0  # Minimum distance between peaks (vs 7.0 for H-NMR)

print("Peak Detection Parameters:")
print(f"  Height threshold: {C13_HEIGHT_FRAC*100:.1f}% of max intensity")
print(f"  Prominence threshold: {C13_PROMINENCE_FRAC*100:.2f}% of max intensity")
print(f"  Min distance: {C13_MIN_DISTANCE_HZ} Hz")
print("\\nüìç Note: Lowered thresholds to detect weaker peaks (e.g., near Œ¥ ~0 ppm)")

# Peak detection using positive frequencies only
res = plot_full_carbon_and_zoom_with_peaks(
    positive_frequencies, 
    positive_magnitude,
    title=f"{name} - Carbon-13 NMR Spectrum{' (Denoised)' if USE_DENOISING else ''}",
    spectrometer_freq=C13_SPECTROMETER_FREQ,
    ppm_range=220,  # Carbon-13 typically spans 0-220 ppm
    identify_functional_groups=nmr.identify_functional_groups,
    ppm_shifts=C13_PPM_SHIFTS,
    min_distance_hz=C13_MIN_DISTANCE_HZ,
    height_frac=C13_HEIGHT_FRAC,
    prominence_frac=C13_PROMINENCE_FRAC,
    show=False  # suppress visualization in this cell
)

identified_groups = res.get("identified_groups", [])
peaks = res.get("peaks", np.array([]))
ppm_axis_res = res.get("ppm_axis", ppm_axis)
intensity = res.get("intensity", positive_magnitude)

print(f"\nDetected {len(peaks)} peaks in the spectrum")
if len(peaks) > 0:
    peak_ppms_list = [ppm_axis_res[p] for p in peaks]
    print(f"Peak positions (ppm): {[f'{p:.2f}' for p in peak_ppms_list]}")
    if any(p < 2.0 for p in peak_ppms_list):
        print("‚úÖ Peak(s) detected in TMS/low-field region (Œ¥ < 2 ppm)")
````

# Peak Integration

Calculate the integrated area under each detected peak to estimate relative carbon counts (though in practice, Carbon-13 NMR integration is less reliable than proton NMR due to relaxation effects).

````python
from scipy.integrate import simpson

# Integrate each peak and visualize
peak_data = []
integration_window_ppm = 2.0  # Window around each peak (¬±1 ppm)

if len(peaks) > 0:
    for i, peak_idx in enumerate(peaks):
        peak_ppm = ppm_axis_res[peak_idx]
        peak_height = intensity[peak_idx]
        
        # Define integration window
        left_ppm = peak_ppm - integration_window_ppm
        right_ppm = peak_ppm + integration_window_ppm
        
        # Find indices in window
        mask = (ppm_axis_res >= left_ppm) & (ppm_axis_res <= right_ppm)
        window_ppm = ppm_axis_res[mask]
        window_intensity = intensity[mask]
        
        # Integrate using Simpson's rule
        if len(window_intensity) > 2:
            integral = simpson(window_intensity, x=window_ppm)
        else:
            integral = np.trapz(window_intensity, x=window_ppm)
        
        peak_data.append({
            'peak_number': i + 1,
            'ppm': peak_ppm,
            'height': peak_height,
            'integral': abs(integral),
            'window_ppm': window_ppm,
            'window_intensity': window_intensity
        })
    
    # Normalize integrals to the smallest peak (for relative carbon counts)
    if peak_data:
        min_integral = min(p['integral'] for p in peak_data if p['integral'] > 0)
        for p in peak_data:
            p['relative_integral'] = p['integral'] / min_integral if min_integral > 0 else 0
            p['relative_integral_int'] = int(round(p['relative_integral']))
    
    # Display integration results with visualization
    print("\nPeak Integration Results:")
    print("=" * 90)
    print(f"{'Peak':<6} {'Œ¥ (ppm)':<10} {'Height (A.U.)':<15} {'Integral':<12} {'# Carbons':<12}")
    print("-" * 90)
    for p in peak_data:
        print(f"{p['peak_number']:<6} {p['ppm']:<10.2f} {p['height']:<15.1f} "
              f"{p['integral']:<12.2e} {p['relative_integral_int']:<12}")
    
    # Visualize integrations (style closer to H-NMR cell 42)
    print("\nüìä Visualizing Peak Integrations:")
    fig, axes = plt.subplots(1, len(peak_data), figsize=(4*len(peak_data), 4))
    if len(peak_data) == 1:
        axes = [axes]
    
    for idx, p in enumerate(peak_data):
        ax = axes[idx]
        num_carbons = p['relative_integral_int']
        
        # Plot spectrum in integration window
        ax.plot(p['window_ppm'], p['window_intensity'], linewidth=2, color='#1f77b4')
        ax.fill_between(p['window_ppm'], p['window_intensity'], alpha=0.3, color='#1f77b4')
        
        # Peak position marker
        ax.axvline(p['ppm'], color='red', linestyle='-', linewidth=2, alpha=0.7, label=f"Peak {p['peak_number']}")
        
        # Badge: number of carbons
        ax.text(p['ppm'], max(p['window_intensity']) * 0.92, f"{num_carbons} C", 
                ha='center', fontsize=11, fontweight='bold',
                bbox=dict(boxstyle='round,pad=0.5', facecolor='yellow', edgecolor='black', alpha=0.85))
        
        # Integration label
        ax.text(p['window_ppm'][0] + (p['window_ppm'][-1] - p['window_ppm'][0]) * 0.04, max(p['window_intensity']) * 0.80,
                f"‚à´ = {p['relative_integral']:.2f}", fontsize=10, fontweight='bold', color='darkblue',
                bbox=dict(boxstyle='round,pad=0.4', facecolor='lightblue', alpha=0.7))
        
        ax.invert_xaxis()
        ax.set_xlabel('Chemical Shift Œ¥ (ppm)', fontsize=10, fontweight='bold')
        ax.set_ylabel('Intensity (A.U.)', fontsize=10, fontweight='bold')
        ax.set_title(f"Peak {p['peak_number']}: Œ¥ = {p['ppm']:.2f} ppm", fontsize=11, fontweight='bold')
        ax.grid(alpha=0.3, linestyle=':')
        ax.legend(fontsize=9)
    
    plt.tight_layout()
    plt.show()
    
    print("\n‚ö†Ô∏è  Note: Carbon-13 integration is less reliable than ¬πH-NMR due to:")
    print("   - Variable relaxation times (T1) for different carbon types")
    print("   - NOE effects in proton-decoupled spectra")
    print("   - Use primarily for qualitative comparison, not quantitative analysis")
else:
    print("No peaks detected for integration")
````

# J-Coupling Analysis (Carbon-13)

**Important Note about Carbon-13 J-Coupling:**

In **proton-decoupled Carbon-13 NMR** (most common):
- $^{13}C$-$^{1}H$ coupling is **removed** by broadband decoupling
- $^{13}C$-$^{13}C$ coupling is **not observed** (only ~0.01% chance of adjacent $^{13}C$ atoms)
- Result: Each unique carbon shows as a **single sharp peak** (singlet)

In **coupled Carbon-13 NMR** (less common):
- $^{13}C$-$^{1}H$ coupling is visible
- Can determine number of attached hydrogens (DEPT technique is better for this)
- Typical $^{1}J_{CH}$ values: 125-250 Hz

If your spectrum shows multiplets, it may be:
1. A coupled $^{13}C$ spectrum (not decoupled)
2. Satellite peaks from $^{13}C$-$^{13}C$ coupling in enriched samples
3. Incomplete decoupling

Let's check for any multiplet structure:

````python
# Check for multiplet structure in each identified peak region
from scipy.ndimage import gaussian_filter1d

def detect_multiplet_c13(ppm_axis, intensity, center_ppm, window=0.5):
    """
    Detect potential multiplet structure in a Carbon-13 peak.
    
    Args:
        ppm_axis: PPM values
        intensity: Spectrum intensity
        center_ppm: Center of peak to analyze
        window: Window size in ppm (¬±window around center)
    
    Returns:
        dict with multiplet information
    """
    # Mask for the region of interest
    mask = (ppm_axis > center_ppm - window) & (ppm_axis < center_ppm + window)
    region_ppm = ppm_axis[mask]
    region_intensity = intensity[mask]
    
    if len(region_intensity) < 10:
        return None
    
    # Smooth to reduce noise
    smoothed = gaussian_filter1d(region_intensity, sigma=2)
    
    # Find sub-peaks
    sub_peaks, props = find_peaks(
        smoothed,
        height=0.2 * max(smoothed),
        prominence=0.1 * max(smoothed),
        distance=5
    )
    
    if len(sub_peaks) <= 1:
        return {
            'type': 'singlet',
            'num_peaks': 1,
            'j_coupling': None,
            'center_ppm': center_ppm
        }
    
    # Calculate J-coupling from peak separations
    sub_ppms = region_ppm[sub_peaks]
    sub_ppms_sorted = np.sort(sub_ppms)
    
    # Convert ppm differences to Hz
    j_couplings_hz = []
    for i in range(len(sub_ppms_sorted) - 1):
        delta_ppm = sub_ppms_sorted[i+1] - sub_ppms_sorted[i]
        delta_hz = delta_ppm * C13_SPECTROMETER_FREQ
        j_couplings_hz.append(delta_hz)
    
    # Determine multiplet type
    n_peaks = len(sub_peaks)
    if n_peaks == 2:
        multiplet_type = 'doublet'
    elif n_peaks == 3:
        multiplet_type = 'triplet'
    elif n_peaks == 4:
        multiplet_type = 'quartet'
    else:
        multiplet_type = f'{n_peaks}-plet'
    
    return {
        'type': multiplet_type,
        'num_peaks': n_peaks,
        'j_coupling': np.mean(j_couplings_hz) if j_couplings_hz else None,
        'j_couplings_all': j_couplings_hz,
        'center_ppm': center_ppm,
        'sub_ppms': sub_ppms_sorted.tolist()
    }

# Analyze each identified peak
print("\\nMultiplet Analysis:")
print("=" * 80)

multiplet_results = []
has_coupling = False  # Track if we detect any multiplets

for i, peak_idx in enumerate(peaks):
    peak_ppm = ppm_axis_res[peak_idx]
    result = detect_multiplet_c13(ppm_axis_res, intensity, peak_ppm, window=0.5)
    
    if result:
        multiplet_results.append(result)
        print(f"\\nPeak {i+1} at Œ¥ {peak_ppm:.2f} ppm:")
        print(f"  Type: {result['type']}")
        
        if result['type'] != 'singlet':
            has_coupling = True
            print(f"  J-coupling: {result['j_coupling']:.2f} Hz")
            print(f"  Sub-peaks at: {[f'{p:.2f}' for p in result['sub_ppms']]} ppm")
            print(f"  ‚ö†Ô∏è  Multiplet detected! This may indicate:")
            print(f"     - Coupled ¬π¬≥C spectrum (not fully decoupled)")
            print(f"     - ¬π¬≥C-¬πH coupling visible")
            if result['j_coupling'] > 100:
                print(f"     - Likely ¬πJ(C-H) coupling (~{result['j_coupling']:.0f} Hz)")
        else:
            print(f"  Single peak (typical for decoupled ¬π¬≥C-NMR)")

# Determine if spectrum is coupled or decoupled
if has_coupling:
    print("\\nüî¨ SPECTRUM APPEARS TO BE COUPLED")
    print("   Multiple sub-peaks detected, indicating ¬π¬≥C-¬πH coupling is visible")
    print("\\nüìö About Coupled vs Decoupled Spectra:")
    print("   ‚Ä¢ Decoupled: Broadband proton decoupling removes ¬π¬≥C-¬πH coupling ‚Üí singlets")
    print("   ‚Ä¢ Coupled: No decoupling ‚Üí multiplets reveal # of attached hydrogens")
    print("\\nüõ†Ô∏è  How to Decouple (Experimental):")
    print("   1. Apply broadband ¬πH decoupling during acquisition (hardware)")
    print("   2. Use WALTZ-16 or GARP decoupling sequences")
    print("   3. Computational: Deep learning techniques can simulate decoupling")
    print("\\nüí° Visualizing Multiplet Patterns:")
else:
    print("\\n‚úÖ SPECTRUM APPEARS TO BE DECOUPLED")
    print("   All peaks appear as singlets (expected for standard ¬π¬≥C-NMR)")
  
````

#### **Step 2: Fourier Transform (Time ‚Üí Frequency)**

$$
S[k] = \sum_{n=0}^{N-1} \text{FID}[n] \cdot e^{-i2\pi kn/N}
$$

**Converts:**
- Time-domain FID ‚Üí Frequency-domain spectrum
- Complex exponentials ‚Üí Peak positions (chemical shifts)
- T‚ÇÇ decay ‚Üí Lorentzian lineshapes

---

#### **Step 3: Gaussian Smoothing (Frequency Domain)**

$$
S_{\text{smooth}}[i] = \sum_{j} S[j] \cdot \frac{1}{\sigma\sqrt{2\pi}} e^{-(i-j)^{2}/(2\sigma^{2})}
$$

**Applied when:** Deep learning ŒîSNR < +2 dB

**Effect:**
- Reduces high-frequency noise
- Slight peak broadening (trade-off)
- Improved visual clarity

---

#### **Step 4: Chemical Shift Calibration**

$$
\delta_{\text{ppm}} = \frac{\nu_{\text{Hz}}}{\nu_{\text{spectrometer}}} = \frac{\nu_{\text{Hz}}}{100.52 \text{ MHz}}

**Reference:** TMS (tetramethylsilane) at Œ¥ = 0 ppm

Energy splitting for spin-¬Ω nuclei in magnetic field $B_{0}$.

$$

$$
\nu = \nu_{0} (1 - \sigma) \quad \text{where } \sigma = \text{shielding constant}

**4. Relaxation (T‚ÇÇ decay):**

$$
M_{xy}(t) = M_{0} e^{-t/T_{2}} e^{i\omega_{0} t}
