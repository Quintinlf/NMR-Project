## üß† Load Deep Learning Denoising Model

We use a trained deep learning model to enhance signal quality before analysis.

```python
# Define the denoiser architecture
class DilatedResBlock(nn.Module):
    def __init__(self, channels: int, dilation: int = 1, k: int = 11):
        super().__init__()
        pad = dilation * ((k - 1) // 2)
        self.conv1 = nn.Conv1d(channels, channels, k, padding=pad, dilation=dilation)
        self.conv2 = nn.Conv1d(channels, channels, k, padding=pad, dilation=dilation)
    def forward(self, x):
        h = F.relu(self.conv1(x))
        h = self.conv2(h)
        return F.relu(x + h)

class DenoiseNetPhysics(nn.Module):
    def __init__(self, in_ch=2, hidden=64, k=11, dilations=(1,2,4,8,16,32)):
        super().__init__()
        self.inp = nn.Conv1d(in_ch, hidden, k, padding=(k-1)//2)
        self.blocks = nn.Sequential(*[DilatedResBlock(hidden, d, k) for d in dilations])
        self.out = nn.Conv1d(hidden, in_ch, k, padding=(k-1)//2)
    def forward(self, x):
        h = F.relu(self.inp(x))
        h = self.blocks(h)
        return x + self.out(h)

# Load model
denoiser_model = DenoiseNetPhysics().to(device)

# Try loading checkpoint
checkpoint_dir = Path("checkpoints")
best_models = ["DenoiseNetPhysics_final_best.pth", "DenoiseNetPhysics_aggressive_best.pth",
               "DenoiseNetPhysics_extended_best.pth", "DenoiseNetPhysics_finetuned_best.pth"]

loaded = False
for ckpt_name in best_models:
    ckpt_path = checkpoint_dir / ckpt_name
    if ckpt_path.exists():
        try:
            ckpt = torch.load(ckpt_path, map_location=device, weights_only=False)
            denoiser_model.load_state_dict(ckpt['model_state'] if 'model_state' in ckpt else ckpt)
            print(f"‚úÖ Loaded: {ckpt_name}")
            loaded = True
            break
        except: 
            continue

if loaded:
    denoiser_model.eval()
    print("‚úÖ Denoising model ready!")
else:
    print("‚ö†Ô∏è  No model loaded - will show original data only")
```

## üìä Helper Functions for Analysis

```python
def apply_denoising(data, model):
    """Apply deep learning denoising to FID data."""
    data_denoised = data.copy()
    real_part = data[:, 1].astype(np.float32)
    imag_part = data[:, 2].astype(np.float32)
    sig = np.stack([real_part, imag_part])
    
    # Normalize
    L = sig.shape[1]
    late = slice(int(0.8*L), L)
    baseline = np.median(sig[:, late], axis=1, keepdims=True)
    sig_norm = sig - baseline
    
    # Denoise
    with torch.no_grad():
        x = torch.from_numpy(sig_norm).unsqueeze(0).to(device)
        y = model(x).cpu().squeeze(0).numpy()
    
    # Update data
    data_denoised[:, 1] = (y[0] + baseline[0, 0])
    data_denoised[:, 2] = (y[1] + baseline[1, 0])
    
    return data_denoised

def calc_snr(data):
    """Calculate SNR of FID data."""
    env = np.sqrt(data[:,1]**2 + data[:,2]**2)
    s = int(0.2*len(env))
    signal = np.sqrt(np.mean(env[:s]**2))
    noise = np.sqrt(np.mean(env[-s:]**2)) + 1e-12
    return 20*np.log10(signal / noise)

def plot_fid_comparison(data_original, data_denoised, title, snr_before, snr_after):
    """Plot FID before and after denoising side by side."""
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(16, 5))
    
    # Before denoising
    time = data_original[:, 0]
    ax1.plot(time, data_original[:, 1], label='Real', alpha=0.7, linewidth=1)
    ax1.plot(time, data_original[:, 2], label='Imaginary', alpha=0.7, linewidth=1)
    ax1.set_xlabel('Time (s)', fontweight='bold')
    ax1.set_ylabel('Signal Intensity', fontweight='bold')
    ax1.set_title(f'{title} - BEFORE Denoising\nSNR: {snr_before:.1f} dB', fontweight='bold')
    ax1.legend()
    ax1.grid(alpha=0.3)
    
    # After denoising
    ax2.plot(time, data_denoised[:, 1], label='Real', alpha=0.7, linewidth=1, color='green')
    ax2.plot(time, data_denoised[:, 2], label='Imaginary', alpha=0.7, linewidth=1, color='orange')
    ax2.set_xlabel('Time (s)', fontweight='bold')
    ax2.set_ylabel('Signal Intensity', fontweight='bold')
    snr_improvement = snr_after - snr_before
    ax2.set_title(f'{title} - AFTER Denoising\nSNR: {snr_after:.1f} dB ({snr_improvement:+.1f} dB)', 
                  fontweight='bold')
    ax2.legend()
    ax2.grid(alpha=0.3)
    
    plt.tight_layout()
    plt.show()
    
    return snr_improvement

print("‚úÖ Helper functions loaded!")
```

```python
def plot_functional_group_zooms(ppm_axis, intensity, peaks, ppm_shifts, 
                                 spectrometer_freq, title_prefix="", window=0.5):
    """
    Create individual zoomed plots for each functional group region with labeled peaks.
    
    Args:
        ppm_axis: Chemical shift axis in ppm
        intensity: Spectrum intensity
        peaks: Detected peak indices
        ppm_shifts: Dictionary of functional group ppm ranges
        spectrometer_freq: Spectrometer frequency in MHz
        title_prefix: Prefix for plot titles
        window: Window size around each peak group in ppm
    """
    if len(peaks) == 0:
        print("‚ö†Ô∏è  No peaks detected for functional group zoom plots")
        return
    
    # Convert peak indices to ppm values
    peak_ppms = ppm_axis[peaks]
    peak_heights = intensity[peaks]
    
    # Group peaks by functional group
    group_peaks = {}
    for peak_idx, peak_ppm in zip(peaks, peak_ppms):
        for group_name, ppm_range_str in ppm_shifts.items():
            # Handle negative ranges like "-0.5-0.5ppm" or "5-50ppm"
            range_clean = ppm_range_str.replace("ppm", "").strip()
            # Split carefully to handle negative numbers
            if range_clean.startswith('-'):
                parts = range_clean[1:].split('-')
                ppm_min = -float(parts[0])
                ppm_max = float(parts[1])
            else:
                parts = range_clean.split('-')
                ppm_min = float(parts[0])
                ppm_max = float(parts[1])
            
            if ppm_min <= peak_ppm <= ppm_max:
                if group_name not in group_peaks:
                    group_peaks[group_name] = []
                group_peaks[group_name].append((peak_ppm, peak_idx))
                break
    
    # Create zoom plot for each functional group with peaks
    for group_name, peaks_in_group in group_peaks.items():
        if len(peaks_in_group) == 0:
            continue
            
        # Determine zoom window
        group_ppms = [p[0] for p in peaks_in_group]
        zoom_min = min(group_ppms) - window
        zoom_max = max(group_ppms) + window
        
        # Create mask for zoom region
        mask = (ppm_axis >= zoom_min) & (ppm_axis <= zoom_max)
        
        # Plot
        fig, ax = plt.subplots(figsize=(10, 5))
        ax.plot(ppm_axis[mask], intensity[mask], linewidth=2, color='#1f77b4', label='Spectrum')
        
        # Mark each peak with vertical line and label
        for peak_ppm, peak_idx in peaks_in_group:
            peak_height = intensity[peak_idx]
            
            # Vertical line at peak
            ax.axvline(peak_ppm, color='red', linestyle='--', linewidth=1.5, alpha=0.7)
            
            # Label above peak
            ax.annotate(
                f'{peak_ppm:.2f} ppm',
                xy=(peak_ppm, peak_height),
                xytext=(0, 20),
                textcoords='offset points',
                ha='center',
                fontsize=11,
                fontweight='bold',
                color='darkred',
                bbox=dict(boxstyle='round,pad=0.5', facecolor='yellow', 
                         edgecolor='darkred', alpha=0.8),
                arrowprops=dict(arrowstyle='->', color='darkred', lw=1.5)
            )
        
        ax.invert_xaxis()
        ax.set_xlabel('Chemical Shift Œ¥ (ppm)', fontsize=13, fontweight='bold')
        ax.set_ylabel('Intensity (A.U.)', fontsize=13, fontweight='bold')
        
        # Remove \n and simplify title
        title_clean = title_prefix.replace('\\n', ' - ').strip()
        if title_clean.endswith(' - '):
            title_clean = title_clean[:-3]
        ax.set_title(f'{title_clean}{group_name} (Œ¥ {zoom_min:.2f} - {zoom_max:.2f} ppm)', 
                    fontsize=13, fontweight='bold', pad=15)
        ax.legend(fontsize=11)
        ax.grid(alpha=0.3, linestyle=':', linewidth=0.8)
        
        # Add text box with functional group info - positioned to avoid title overlap
        info_text = f"Functional Group: {group_name}\nPeaks detected: {len(peaks_in_group)}"
        info_text += f"\nRegion: {ppm_shifts[group_name]}"
        
        ax.text(0.02, 0.85, info_text,
               transform=ax.transAxes,
               fontsize=10,
               verticalalignment='top',
               bbox=dict(boxstyle='round', facecolor='lightblue', alpha=0.7))
        
        plt.tight_layout()
        plt.show()
        
        # Print peak positions like in cell 39
        peak_positions_str = ', '.join([f'{p[0]:.2f}' for p in peaks_in_group])
        print(f"  {group_name}: {peak_positions_str} ppm")
    
    print(f"\n‚úÖ Created {len(group_peaks)} functional group zoom plots")

print("‚úÖ Enhanced functional group zoom plotting function loaded!")
```

# üìà Part 1: Hydrogen-1 NMR Analysis

## ¬πH-NMR Spectroscopy Overview

**Hydrogen-1 NMR** is the most common NMR technique for organic structure determination:
- Detects all hydrogen atoms in the molecule
- Chemical shift range: **0-12 ppm** (relative to TMS)
- Provides information about:
  - Number of unique hydrogen environments
  - Relative number of hydrogens (integration)
  - Neighboring hydrogens (J-coupling/splitting patterns)

### Chemical Shift Ranges (¬πH):
| Type | Œ¥ (ppm) | Example |
|------|---------|---------|
| Alkyl (CH‚ÇÉ, CH‚ÇÇ) | 0.5-2.0 | Saturated hydrocarbons |
| Allylic/Œ± to C=O | 2.0-3.0 | CH‚ÇÇ next to double bond |
| C-O (alcohols, ethers) | 3.0-5.0 | -OCH‚ÇÉ, -OCH‚ÇÇ- |
| Alkene (=C-H) | 5.0-6.0 | C=C-H |
| Aromatic | 6.5-8.5 | Benzene rings |
| Aldehyde | 9.0-10.0 | -CHO |
| Carboxylic acid | 10-13 | -COOH |

# üìä Part 2: Carbon-13 NMR Analysis

## ¬π¬≥C-NMR Spectroscopy Overview

**Carbon-13 NMR** provides the carbon framework of molecules:
- Natural abundance: ~1.1% (vs. 99.98% for ¬πH)
- Chemical shift range: **0-220 ppm** (much wider than ¬πH)
- Usually **proton-decoupled** ‚Üí single peak per unique carbon
- Less sensitive than ¬πH-NMR but gives complementary structural information

### Chemical Shift Ranges (¬π¬≥C):
| Carbon Type | Œ¥ (ppm) | Example |
|-------------|---------|---------|
| Alkyl (sp¬≥) | 0-50 | CH‚ÇÉ, CH‚ÇÇ, CH |
| C-O, C-N | 50-90 | Alcohols, ethers, amines |
| Alkyne (sp) | 65-85 | C‚â°C |
| Alkene (sp¬≤) | 100-150 | C=C |
| Aromatic | 110-160 | Benzene rings |
| Carbonyl | 160-220 | C=O (ketones, esters, acids) |

### Analysis Features:
- **FID Before/After Denoising** comparison
- **Spectrum Before/After Gaussian Filtering** comparison
- Peak detection and functional group identification

# üî¨ Part 3: 2D NMR Spectroscopy

## 2D NMR Overview

**Two-dimensional NMR** techniques provide correlation information between nuclei:

### COSY (Correlation Spectroscopy)
- **¬πH-¬πH** correlation through **J-coupling**
- Shows which protons are **connected** (2-3 bonds apart)
- Cross-peaks indicate **neighboring hydrogens**
- Applications: Determine spin systems, trace coupling networks

### HMQC (Heteronuclear Multiple Quantum Coherence)
- **¬πH-¬π¬≥C** correlation through **direct (1-bond) J-coupling**
- Shows which **protons are attached** to which **carbons**
- Applications: Assign ¬πH and ¬π¬≥C peaks, determine CH, CH‚ÇÇ, CH‚ÇÉ

### HMBC (Heteronuclear Multiple Bond Correlation)
- **¬πH-¬π¬≥C** correlation through **long-range (2-4 bond) J-coupling**
- Shows which protons are **near** (but not attached to) which carbons
- Applications: Determine carbon skeleton connectivity, identify quaternary carbons
