```python
from rdkit import Chem
from rdkit.Chem import Draw, AllChem, Descriptors
import py3Dmol
from rdkit.Chem import rdMolDescriptors
import pubchempy as pcp
import requests
```

# Compounds to Analyze

- Valerenic acid
- Valerenal
- Acetoxyvalerenic acid
- Acetoxyvalerenal 
- Hydroxyvalerenic acid
- Hydroxyvalerenal

```python
# List of compounds to analyze
compounds_to_analyze = [
    "valerenic acid",
    "valerenal",
    "Acetoxyvalerenic acid",
    "Acetoxyvalerenal",
    "Hydroxyvalerenic acid",
    "hydroxyvalerenal"
]

inhibitor_dict = {
    "thc": "CYP2C9, CYP3A4 inhibitor",
    "cbd": "CYP2C19, CYP3A4 inhibitor",
    "cbn": "CYP2C9 inhibitor",
    "5-methoxy-alpha-methyltryptamine": "No major CYP inhibition reported",
    "bupropion": "CYP2D6 inhibitor"
}

# Loop through each compound
for compound_name in compounds_to_analyze:
    print("\n" + "="*80)
    print(f"{'ANALYZING: ' + compound_name.upper():^80}")
    print("="*80 + "\n")
    
    try:
        compound = pcp.get_compounds(compound_name, 'name')[0]
        inchi = compound.inchi
        mol = Chem.MolFromInchi(inchi)

        chemical_name = compound_name
        inhibitor_info = inhibitor_dict.get(chemical_name.lower(), "Unknown")

        header = f" Compound Information: {compound_name.capitalize()} "
        print(f"\n{'='*len(header)}")
        print(header)
        print(f"{'='*len(header)}")

        print(f"{'InChI'.ljust(20)}: {inchi}")
        print(f"{'LogP (polarity)'.ljust(20)}: {Descriptors.MolLogP(mol):.2f}")
        print(f"{'TPSA (Å²)'.ljust(20)}: {Descriptors.TPSA(mol):.2f}")
        print(f"{'H-Bond Donors'.ljust(20)}: {Descriptors.NumHDonors(mol)}")
        print(f"{'H-Bond Acceptors'.ljust(20)}: {Descriptors.NumHAcceptors(mol)}")
        print(f"{'Rotatable Bonds'.ljust(20)}: {Descriptors.NumRotatableBonds(mol)}")
        print(f"{'Inhibitor Profile'.ljust(20)}: {inhibitor_info}")

        print(f"{'='*len(header)}\n")

        # Get SMILES and generate visualizations
        smiles = compound.isomeric_smiles
        if smiles:
            mol = Chem.MolFromSmiles(smiles)
        else:
            print("SMILES not found, using InChI to generate molecule.")
            mol = Chem.MolFromInchi(inchi)

        if mol:
            # 2D Structure
            AllChem.Compute2DCoords(mol)
            img_2d = Draw.MolToImage(mol, size=(300, 300))
            display(img_2d)

            # 3D Structure
            mol_3d = Chem.AddHs(mol)
            AllChem.EmbedMolecule(mol_3d, AllChem.ETKDG())
            AllChem.UFFOptimizeMolecule(mol_3d)
            mb = Chem.MolToMolBlock(mol_3d)
            view = py3Dmol.view(width=400, height=400)
            view.addModel(mb, 'mol')
            view.setStyle({'stick':{}})
            view.zoomTo()
            view.show()
        else:
            print("Could not generate molecule from SMILES or InChI.")

        # Bioassay information
        cid = compound.cid
        url = f"https://pubchem.ncbi.nlm.nih.gov/rest/pug/compound/cid/{cid}/aids/TXT"
        response = requests.get(url)
        aids = response.text.strip().split('\n')[:50]

        for aid in aids:
            desc_url = f"https://pubchem.ncbi.nlm.nih.gov/rest/pug/assay/aid/{aid}/description/JSON"
            desc_resp = requests.get(desc_url)
            desc = desc_resp.json()
            description = ""
            try:
                description = desc['AssayDescription']['Assay']['Description']
            except KeyError:
                try:
                    description = desc['PC_AssayContainer'][0]['assay']['descr']['description'][0]
                except Exception:
                    continue
            if any(word in description.lower() for word in ["inhibit", "cyp", "enzyme"]):
                link = f"https://pubchem.ncbi.nlm.nih.gov/bioassay/{aid}"
                print("="*60)
                print(f"AID: {aid}")
                print(f"Description: {description}")
                print(f"Link: {link}\n")
    
    except Exception as e:
        print(f"Error processing {compound_name}: {str(e)}")
        continue

print("\n" + "="*80)
print("ANALYSIS COMPLETE")
print("="*80)
```
