```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from pathlib import Path
from importlib import reload
from functions import nmr_function as nmr

reload(nmr)

# Load ALL files from flashdrive_data folder
folder = Path("flashdrive_data")

# List all ASC files
asc_files = sorted([f for f in folder.glob("*.asc")])
total_files = len(asc_files)

print(f"ðŸ“‚ Found {total_files} ASC files in {folder}/")
print("="*70)
print("FILE LIST:")
print("="*70)

# Print ALL files, no truncation
for i, f in enumerate(asc_files, 1):
    print(f"  {i:2d}. {f.name}")

print("\n" + "="*70)
print("LOADING AND PLOTTING ALL FLASHDRIVE FILES")
print("="*70 + "\n")

# Load and plot each file
for idx, file in enumerate(asc_files, 1):
    try:
        df, name = nmr.load_fid_and_preview(str(file))
        
        # Convert to numpy
        arr = df if isinstance(df, np.ndarray) else df.to_numpy()
        
        print(f"\n{'='*70}")
        print(f"ðŸ“Š File {idx}/{total_files}: {name}")
        print(f"{'='*70}")
        print(f"  Shape:         {arr.shape[0]:,} points Ã— {arr.shape[1]} columns")
        print(f"  Data Type:     {arr.dtype}")
        print(f"  X Range:       {arr[0,0]:.6f} â†’ {arr[-1,0]:.6f}")
        print(f"  Real Range:    {arr[:,1].min():+.6e} â†’ {arr[:,1].max():+.6e}")
        print(f"  Imag Range:    {arr[:,2].min():+.6e} â†’ {arr[:,2].max():+.6e}")
        
        # Plot with all data points
        fig, axes = plt.subplots(3, 1, figsize=(14, 10))
        
        # Plot 1: Real component
        axes[0].plot(arr[:, 0], arr[:, 1], linewidth=0.5, color='blue')
        axes[0].set_title(f'{name} - Real Component (ALL {arr.shape[0]:,} points)', 
                         fontweight='bold', fontsize=12)
        axes[0].set_xlabel('X (Time or Frequency)')
        axes[0].set_ylabel('Real')
        axes[0].grid(True, alpha=0.3)
        
        # Plot 2: Imaginary component
        axes[1].plot(arr[:, 0], arr[:, 2], linewidth=0.5, color='orange')
        axes[1].set_title(f'Imaginary Component (ALL {arr.shape[0]:,} points)', 
                         fontweight='bold', fontsize=12)
        axes[1].set_xlabel('X (Time or Frequency)')
        axes[1].set_ylabel('Imaginary')
        axes[1].grid(True, alpha=0.3)
        
        # Plot 3: Magnitude
        magnitude = np.sqrt(arr[:, 1]**2 + arr[:, 2]**2)
        axes[2].plot(arr[:, 0], magnitude, linewidth=0.5, color='green')
        axes[2].set_title(f'Magnitude (ALL {arr.shape[0]:,} points)', 
                         fontweight='bold', fontsize=12)
        axes[2].set_xlabel('X (Time or Frequency)')
        axes[2].set_ylabel('Magnitude')
        axes[2].grid(True, alpha=0.3)
        
        plt.tight_layout()
        plt.show()
        
    except Exception as e:
        print(f"\nâŒ Error with file {idx}/{total_files} ({file.name}): {e}")
        import traceback
        traceback.print_exc()

print("\n" + "="*70)
print(f"âœ… Finished displaying all {total_files} flashdrive files!")
print("="*70)
```
